```python
troop_movements['is_resistance'] = troop_movements['empire_or_resistance'] == 'resistance'
troop_movements
```

```python
#create a prediction model using sklearn.tree.DecisionTreeClassifier that predicts if a character is joining either the empire or the resistance based on their home world and unit type
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder

le = LabelEncoder()
X = troop_movements['homeworld']
y = le.fit_transform(troop_movements['empire_or_resistance'])

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)

X_train = pd.get_dummies(X_train)
X_test = pd.get_dummies(X_test)

model = DecisionTreeClassifier()

model.fit(X_train, y_train)


```

```python
import seaborn
import matplotlib.pyplot as plt

df = troop_movements

# Group by 'empire_or_resistance' and count the occurrences
counts = df['empire_or_resistance'].value_counts().reset_index()

# Rename columns for clarity
counts.columns = ['faction', 'count']

# Create the bar plot using Seaborn
plt.figure(figsize=(8, 6))
seaborn.barplot(x='faction', y='count', data=counts, palette='viridis')
plt.title('Distribution of Units: Empire vs Resistance')
plt.xlabel('Empire or Resistance')
plt.ylabel('Count')
plt.show()
```

```python
 # create a bar plot that shows feature importance of the model
importances = model.feature_importances_
features = X_train.columns

indices = np.argsort(importances)[::-1]

plt.figure(figsize=(8, 6))
plt.bar(range(len(importances)), importances[indices], align='center')
plt.xticks(range(len(importances)), [features[i] for i in indices], rotation=90)
plt.title('Feature Importance')
plt.show()

```

```python
# save to model as a pickle file named trained_model.pkl
import pickle

with open('trained_model.pkl', 'wb') as f:
    pickle.dump(model, f)


```
